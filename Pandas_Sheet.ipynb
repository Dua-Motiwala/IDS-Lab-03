```python
# Pandas Practice Sheet

import pandas as pd
import numpy as np

# Creating DataFrames

df = pd.DataFrame({
    "a": [4, 5, 6],
    "b": [7, 8, 9],
    "c": [10, 11, 12]
}, index=[1, 2, 3])
print("DataFrame:\n", df)
```

```python
# Subsetting Data

print("Head:\n", df.head(2))
print("Tail:\n", df.tail(2))
print("iloc rows 0-1:\n", df.iloc[0:2])
print("loc column b:\n", df.loc[:, 'b'])
print("Conditional:\n", df[df['a'] > 4])
```

```python
# Sorting / Renaming

print("Sort by b:\n", df.sort_values('b'))
print("Rename column:\n", df.rename(columns={'a': 'alpha'}))
```

```python
# Dropping

print("Drop column b:\n", df.drop(columns=['b']))
```

```python
# Grouping & Aggregation

data = pd.DataFrame({
    'team': ['A','A','B','B','C'],
    'points': [10, 15, 10, 20, 30]
})
print("Group by team:\n", data.groupby('team')['points'].sum())
```

```python
# Missing Data

df2 = pd.DataFrame({
    'col1': [1, 2, np.nan],
    'col2': [4, np.nan, 6]
})
print("Drop NA:\n", df2.dropna())
print("Fill NA:\n", df2.fillna(0))
```

```python
# New Columns

df['sum'] = df['a'] + df['b']
print("With new column:\n", df)
```

```python
# Merging DataFrames

df1 = pd.DataFrame({'x1':[1,2,3], 'x2':['A','B','C']})
df2 = pd.DataFrame({'x1':[1,2,4], 'x3':['T','F','T']})
print("Merge inner:\n", pd.merge(df1, df2, how='inner', on='x1'))
print("Merge outer:\n", pd.merge(df1, df2, how='outer', on='x1'))
```

```python
# Plotting

df['a'].plot(title="Column a Line Plot")
df['a'].plot.bar(title="Column a Bar Plot")
df.plot.scatter(x='a', y='c', title="Scatter Plot")
```
