```python
#Task 1

import numpy as np

matrix = np.random.randint(0,101, (10,10))
print("Original Matrix:", matrix)

row_sum = matrix.sum(axis = 1)
max_row_idx = np.argmax(row_sum)
print("\nRow sums:", row_sum)
print("Row index with highest sum:", max_row_idx)
print("Highest Sum:", matrix[max_row_idx])

min_val = matrix.min()
max_val = matrix.max()
normalized_mat = (matrix - min_val)/(max_val - min_val)
print("Normalized Matrix:", normalized_mat)

row_mean = normalized_mat.mean(axis=1)
transformed_mat = normalized_mat.copy()
np.fill_diagonal(transformed_mat, row_mean)
print("\nTransformed Matrix (row means on diagonal):\n", transformed_mat)

determinant = np.linalg.det(transformed_mat)
print("\nDeterminant of transformed matrix:", determinant)
```

```python
## Task 2

import matplotlib.pyplot as plt

def outliers_1d(arr):
    print("\nArray:", arr)
    mean = np.mean(arr)
    print("Mean:", mean)
    
    std = np.std(arr)
    print("Standard Deviation:", std)
    
    diff = arr - mean
    print("Difference from Mean:", diff)
    
    outliers = np.abs(diff) > std
    print("Outliers:", outliers)

    plt.scatter(range(len(arr)), arr, label="Data")
    plt.scatter(np.where(outliers)[0], arr[outliers], color='red', label="Outliers")
    plt.title("Outlier Detection (1D)")
    plt.legend()
    plt.show()


def outliers_2d(arr):
    print("\nArray:\n", arr)
    mean = np.mean(arr, axis=0)
    print("Column-wise Mean:", mean)
    
    std = np.std(arr, axis=0)
    print("Column-wise Std Dev:", std)
    
    diff = arr - mean
    print("Difference from Mean:\n", diff)
    
    outliers = np.abs(diff) > std
    print("Outliers:\n", outliers)

    for col in range(arr.shape[1]):
        plt.scatter(range(arr.shape[0]), arr[:, col], label=f"Column {col}")
        plt.scatter(np.where(outliers[:, col])[0], arr[outliers[:, col], col],
                    color='red', marker='x', s=100, label=f"Outliers Col {col}")
    
    plt.title("Outlier Detection (2D, Column-wise)")
    plt.xlabel("Row index")
    plt.ylabel("Value")
    plt.legend()
    plt.show()

    
print("1D array")
arr = np.array(list(map(int, input("Enter numbers separated by spaces: ").split())))
outliers_1d(arr)
    
print("2D array: ")
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))
print("Enter elements row by row (space separated):")
arr = []
for i in range(rows):
    row = list(map(int, input().split()))
    arr.append(row)
arr = np.array(arr)
outliers_2d(arr)
```

```python
#Task 3

import numpy as np
import matplotlib.pyplot as plt
 
rolls = 100000
die1 = np.random.randint(1, 7, rolls)
die2 = np.random.randint(1, 7, rolls)
sums = die1 + die2

values, counts = np.unique(sums, return_counts=True)
empirical_prob = counts / rolls

print("Empirical Probability Distribution (2 dice):")
for v, p in zip(values, empirical_prob):
    print(f"Sum {v}: {p:.3f}")

theoretical_prob = {s: 0 for s in range(2, 13)}
for i in range(1, 7):
    for j in range(1, 7):
        theoretical_prob[i + j] += 1
for k in theoretical_prob:
    theoretical_prob[k] /= 36

print("\nTheoretical Probability Distribution (2 dice):")
for v, p in theoretical_prob.items():
    print(f"Sum {v}: {p:.3f}")

plt.bar(values - 0.2, empirical_prob, width=0.4, label="Empirical")
plt.bar(list(theoretical_prob.keys()), list(theoretical_prob.values()), 
        width=0.4, alpha=0.7, label="Theoretical")
plt.xlabel("Dice Sum")
plt.ylabel("Probability")
plt.title("Dice Roll Probability Distribution (2 dice)")
plt.legend()
plt.show()

cumulative_empirical = np.cumsum(empirical_prob)
plt.plot(values, cumulative_empirical, marker='o')
plt.xlabel("Dice Sum")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Probability Distribution (2 dice)")
plt.show()

die3 = np.random.randint(1, 7, rolls)
sums3 = die1 + die2 + die3
values3, counts3 = np.unique(sums3, return_counts=True)
empirical_prob3 = counts3 / rolls

plt.bar(values3, empirical_prob3, color="orange")
plt.xlabel("Dice Sum (3 dice)")
plt.ylabel("Probability")
plt.title("Empirical Probability Distribution (3 dice)")
plt.show()
```

```python
#Task 4

import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")

df["Name"] = ["Student" + str(i+1) for i in range(len(df))]

long_df = pd.melt(
    df,
    id_vars=["Name", "gender"],
    value_vars=["math score", "reading score", "writing score"],
    var_name="Subject",
    value_name="Marks"
)

long_df["Subject"] = long_df["Subject"].str.replace(" score", "", regex=False).str.capitalize()

print("Reshaped Dataset (first 5 rows):\n", long_df.head())

top_scorers = long_df.loc[long_df.groupby("Subject")["Marks"].idxmax()]
print("\nTop Scorer in Each Subject:\n", top_scorers)

avg_gender_subject = long_df.groupby(["Subject", "gender"])["Marks"].mean()
print("\nAverage Marks per Gender per Subject:\n", avg_gender_subject)

pivot_table = long_df.pivot_table(values="Marks", index="Subject", columns="gender", aggfunc="mean")
print("\nPivot Table (Subject vs Gender Avg Marks):\n", pivot_table)

top5_students = long_df.sort_values(by="Marks", ascending=False).head(5)
print("\nTop 5 Students Overall:\n", top5_students)
```

```python
#Task 5

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    "Name": ["Ali", "Sara", "John", "Areesha", "Taha", "Noor", "Bilal", "Saima",
             "Omar", "Hina", "Usman", "Mariam", "Zain", "Mishal", "Ahmed"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F", "M", "F", "M", "F", "M", "F", "M"],
    "Study_Hours": [2.5, 3.0, 6.5, 5.0, 8.0, 4.5, 7.0, 3.5, 6.0, 2.0, 9.0, 4.0, 5.5, 7.5, 8.5],
    "Marks": [45, 78, 88, 55, 92, 60, 73, 50, 95, 40, 85, 67, 70, 82, 100]
}

df = pd.DataFrame(data)
print("Dataset:\n", df)

plt.hist(df["Marks"], bins=8, color="skyblue", edgecolor="black")
plt.title("Marks Distribution (Histogram)")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

plt.boxplot(df["Marks"], vert=False, patch_artist=True)
plt.title("Boxplot of Marks")
plt.xlabel("Marks")
plt.show()

print("\nObservations:")
print("1. The histogram shows marks are spread across different ranges, with some clustering in the 70–90 range.")
print("2. The boxplot suggests that most marks fall within the normal range, but very high (100) may be seen as an outlier.")
```
